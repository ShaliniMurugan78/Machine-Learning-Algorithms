**Handling Machine Values**

```python
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(strategy='median')
df_imputed=pd.DataFrame(imputer.fit_transform(df),columns=df.columns)
```

**Separate Input(x) & Target(y)**

```python
x=df.drop("2urvived",axis=1)
y=df["2urvived"]
```

```python
print(x)
```

```python
print(y)
```

**Splitting Training & Testing Data**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

**Model Fitting**

```python
from sklearn.tree import DecisionTreeClassifier,plot_tree
model=DecisionTreeClassifier(criterion="gini",max_depth=4,min_samples_leaf=10,random_state=42)
model.fit(x_train,y_train)
```

**Making Prediction**

```python
y_pred=model.predict(x_test)
print(y_pred)
```

**Model Evalution**

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
accuracy_score=accuracy_score(y_test,y_pred)
confusion_matrix=confusion_matrix(y_test,y_pred)
classification_report=classification_report(y_test,y_pred)
print("accuracy_score",accuracy_score)
print("confusion_matrix",confusion_matrix)
print("classification_report",classification_report)
```

**Visualization**

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20,10))
plot_tree(model, filled=True, feature_names=x.columns.tolist(), class_names=['Not Survived', 'Survived'], rounded=True)
plt.title('Decision Tree Visualization')
plt.show()
```

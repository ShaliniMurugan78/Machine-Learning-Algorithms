```python
#storing all non null values into transactions
transactions=[]
for i in range(0,len(df)):
  transactions.append([str(df.values[i,j]) for j in range (0,20) if str(df.values[i,j])!='0'])
```

```python
!pip install apyori
```

```python
from apyori import apriori
```

```python
rules=apriori(transactions=transactions,min_support=0.003,min_confidence=0.2,min_lift=3,min_length=2,max_length=2)
```

```python
results=list(rules)
```

```python
display(results)
```

```python
def inspect(results):
    lhs         = [tuple(rule[2][0][0]) for rule in results]
    rhs         = [tuple(rule[2][0][1]) for rule in results]
    supports    = [rule[1] for rule in results]
    confidences = [rule[2][0][2] for rule in results]
    lifts       = [rule[2][0][3] for rule in results]
    return list(zip(lhs, rhs, supports, confidences, lifts))
```

```python
results_df = pd.DataFrame(inspect(results), columns = ['Left Hand Side', 'Right Hand Side', 'Support', 'Confidence', 'Lift'])
display(results_df)
```
